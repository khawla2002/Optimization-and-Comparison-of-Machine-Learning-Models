# Supervised Learning

### spliting dataset into test set and train set

We shall split the data into the training and the test set. In this the test_size=0.25 denotes that 25% of the data will be kept as the Test set and the remaining 75% will be used for training as the Training set. 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

### feature scaling
This step of feature scaling is an additional step that can increase the speed of machine learning algorithms as we scale down the values of X to a smaller range.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) # fit_transform is only used on the training data
X_test = sc.transform(X_test)
```

# Support Vector Machine Classifier

Once the training test is ready, we can import LinearSVC from sklearn.svm and fit the training set to our model.

LinearSVC is similar to SVC with parameter kernel=’linear’, but implemented in a slightey different way, so it has more flexibility in the choice of penalties and loss functions and should scale better to large numbers of samples.
The classifier.fit() function is then used to train the model.

```python
from sklearn.svm import LinearSVC
import time

start = time.time()   #computing training time
SVC = LinearSVC(C=0.0001)
SVC.fit(X_train, y_train)  # training the model
end = time.time()
print("Training time: ", end - start, "seconds")
print("Score on train data: ", SVC.score(X_train, y_train))
print("Score on test data: ", SVC.score(X_test, y_test))



#this is gonna be useful later in the model performance comparison 
Models = []
scores_test, scores_train = [], []
train_time = []
train_time.append(end - start)
Models.append('SVC')
scores_test.append(SVC.score(X_test, y_test))
scores_train.append(SVC.score(X_train, y_train))
svm=SVC
```

From the above results, we infer that, out of 100 test set data, 69 were correctly classified and 31 were incorrectly classified, leaving us with an accuracy of 69% which is pretty low. Let's see if we can increase it with parameter tuning. 



### Attempt to Increase the accuracy score with gridsearch parameter tuning: 

We know that machine learning algorithms are driven by parameters. These parameters majorly influence the outcome of learning process. The objective of parameter tuning is to find the optimum value for each parameter to improve the accuracy of the model.

But, it is not necessary that higher accuracy models always perform better (for unseen data points). Sometimes, the improvement in model’s accuracy can be due to over-fitting too. To answer this question, we will use cross validation and grid search for parameter tuning on the training data to see how our resulting outputs differs from the output found using the base model above.

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.00001, 0.0001, 0.0005],
                      'dual': (True, False), 'random_state': [123]}

gs_model = GridSearchCV(estimator=LinearSVC(), verbose=1, param_grid=param_grid,
                            scoring='roc_auc', n_jobs=-1)
gs_model.fit(X, y)
```

```python
print(gs_model.best_estimator_)
print(gs_model.best_score_)
```

The results of our more optimal model outperform our initial model with an accuracy score of 0.74 compared to 0.69 prior. 

The one drawback experienced while incorporating GridSearchCV was the runtime. Cross validation & grid tuning lead to longer training times given the repeated number of iterations a model must train through.


# random forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

start = time.time() # computing training time
RandomForest = RandomForestClassifier(n_estimators=25, max_depth=10)
RandomForest.fit(X_train, y_train) # training the model
end = time.time()
print("Training time: ", end - start, "seconds")
print("Accuracy on train data: ", RandomForest.score(X_train, y_train))
print("Accuracy on test data: ", RandomForest.score(X_test, y_test))



# also useful later
train_time.append(end - start)
Models.append('RandomForest')
scores_test.append(RandomForest.score(X_test, y_test))
scores_train.append(RandomForest.score(X_train, y_train))
rfc=RandomForest
```

From the above results, we infer that, out of 100 test set data, 74 were correctly classified and 26 were incorrectly classified, leaving us with an accuracy of 74% which is a better score than our initial SVC (69% before parameter tuning). 

Let's see if it can do better than the optimized SVC with parameter tuning.

### Attempt to Increase the accuracy score with gridsearch parameter tuning: 

```python
max_depth=[2, 8, 16]
n_estimators = [64, 128, 256]
param_grid = dict(max_depth=max_depth, n_estimators=n_estimators)

# Build the grid search
dfrst = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth)
grid = GridSearchCV(estimator=dfrst, param_grid=param_grid, cv = 5)
grid_results = grid.fit(X_train, y_train)

# Summarize the results in a readable format
print("Best: {0}, using {1}".format(grid_results.cv_results_['mean_test_score'], grid_results.best_params_))
results_df = pd.DataFrame(grid_results.cv_results_)
results_df
```

```python
best = grid_results.best_score_
best
```

The results of our more optimal model outperform our initial model with an accuracy score of 0.79 compared to 0.74 prior.

# Conclusion

We can therefore conclude that RandomForest is the most suitable classifier for this dataset with parameters {'max_depth': 16, 'n_estimators': 256} and an accuracy score of 79% which is greater than the accuracy score 74% of the LinearSVC.
